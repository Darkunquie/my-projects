```python
import speech_recognition as sr
from gtts import gTTS

def convert_audio(audio_file, target_language='en'):
  r = sr.Recognizer()
  with sr.AudioFile(audio_file) as source:
    audio = r.record(source)
  try:
    text = r.recognize_google(audio, language='te-IN')
    print("Transcription:", text)
    tts = gTTS(text=text, lang=target_language)
    tts.save("output.mp3")
    print("Audio saved to output.mp3")
  except sr.UnknownValueError:
    print("Could not understand audio")
  except sr.RequestError as e:
    print(f"Could not request results; {e}")
    return
convert_audio("/content/PK_4_V4_S6.wav", target_language='en')
```

```python
import speech_recognition as sr
from gtts import gTTS
from deep_translator import GoogleTranslator
import os
import re

def convert_telugu_audio(audio_file):
    r = sr.Recognizer()
    with sr.AudioFile(audio_file) as source:
        audio = r.record(source)

    try:
        # Recognize Telugu speech using Google Speech Recognition
        text = r.recognize_google(audio, language='te-IN')

        print("Transcription (Telugu):", text)

    except sr.UnknownValueError:
        print("Could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results; {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Directory containing audio files
audio_dir = "/content/drive/MyDrive/PK PROJECT DATA/PK PROJECT"  # Replace with your directory path

# Get a list of audio files and sort based on numerical prefix
audio_files = [
    f for f in os.listdir(audio_dir) if f.endswith((".wav", ".mp3", ".flac"))
]

# Function to extract numerical prefix for sorting
def get_prefix(filename):
    match = re.search(r"^\d+", filename)  # Match digits at the beginning of the filename
    return int(match.group(0)) if match else float('inf')  # Return infinity if no prefix

# Sort files based on the extracted prefix
audio_files.sort(key=get_prefix)

# Iterate through the sorted files
for filename in audio_files:
    filepath = os.path.join(audio_dir, filename)
    print(f"Processing file: {filename}")
    convert_telugu_audio(filepath)
```
